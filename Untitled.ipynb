```python
df = pd.read_csv('dataset/28_1800.csv')
table=df.pivot_table(index=["Name"])
table=table.reset_index()
# features order:
# 0.Name, 1.AirTemp., 2.Alt, 3.DewPoint, 4.Lat, 5.Longt, 6.Pres., 7.R.Humidity, 8.Visib., 9.WindDir., 10.WindGust, 11.WindSpeed

# change the order of features
table2=table.iloc[:,[1,2,3,6,7,8,9,10,11,4,5]]
```

```python
from tqdm.notebook import tqdm
for i in tqdm(range(100)):
    W=[[0.0 for i in range(x)] for j in range(x)]
    for i in range(x):
        for j in range(i+1,x,1):
            z=table2.iloc[i:i+1,:]+table.iloc[j:j+1,:]
            print(z)
            z=z.dropna(axis=1)
            print(z)
            if z.shape[1] == 0:
                continue
            print("Hello")
            W[i][j]=W[j][i]=pairwise_distances(z.values,metric="l1")
            print("i,j=",W[i][j])
            
print(W)
W.to_csv('W.csv', index=False)
```

```python
vectorizer = np.vectorize(lambda x: 1 if x < 15 else 0)
W = np.vectorize(vectorizer)(W)
print(W)
print(len(W),len(W[0]))
```

```python
# degree matrix
D = np.diag(np.sum(np.array(W), axis=1))
print('degree matrix:')
print(D)
# laplacian matrix
L = D - W
print('laplacian matrix:')
print(L)
```

```python
e, v = np.linalg.eig(L)
# eigenvalues
print('eigenvalues:')
print(e)
# eigenvectors
print('eigenvectors:')
print(v)
```

```python
np.where(e<0.1)
```

```python
fig = plt.figure(figsize=[18, 6])
ax1 = plt.subplot(221)
plt.plot(e)
ax1.title.set_text('eigenvalues')
i = np.where(e < 150)[0]
ax2 = plt.subplot(222)
plt.plot(v[:, i[0]])
ax3 = plt.subplot(223)
plt.plot(v[:, i[1]])
ax3.title.set_text('second eigenvector with eigenvalue close to 0')
ax4 = plt.subplot(224)
plt.plot(v[:, i[2]])
ax4.title.set_text('third eigenvector with eigenvalue close to 0')
fig.tight_layout()
```

```python
U = np.array(v[:, i[1]])
U
```

```python

km = KMeans(init='k-means++', n_clusters=4)
km.fit(U.reshape(-1, 1))
km.labels_
```

```python
len(km.labels_)
```

```python
table3=table3.iloc[:500,:]
table3["Cluster"]=km.labels_
#table3

# centroids = clusters' center points
#centroids = cluster.cluster_centers_
title="RelativeHumidity"
weather_clusters=table3[["Lat","Longt","Cluster",title]]
#print(weather_clusters)
weather_clusters.to_csv('ClusteredDatanew.csv', index=False)
```
